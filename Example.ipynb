```python
import labelled_functions as lf

@lf.label
def f(x):
    y = 2*x
    return y

f.show_graph()
```

```python
@lf.label
def g(y, z):
    foo = y*z
    return foo

g.show_graph()
```

```python
(f | g).show_graph()
```

```python
(f | g)(x=2, z=4)
```

```python
@lf.label
def h(z, w=3):
    bar = z + w
    return bar

(f | g | h).show_graph()
```

```python
import pandas as pd
inputs = pd.DataFrame({"x": range(10), "z": range(20, 30)})
inputs
```

```python
lf.pandas_map(f | g | h, inputs)
```
